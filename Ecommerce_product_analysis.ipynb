```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('default')
```

```python
df = pd.read_csv("/content/ecommerce_product_dataset2.csv")

df.head()
```

```python


df.drop_duplicates(inplace=True)

df['DateAdded'] = pd.to_datetime(df['DateAdded'])

df['MonthYear'] = df['DateAdded'].dt.strftime('%b %Y')

df.to_csv("cleaned_ecommerce.csv", index=False)
```

```python
total_sales = df['Sales'].sum()

avg_discount = df['Discount'].mean()

total_products = df['ProductID'].nunique()

avg_rating = df['Rating'].mean()

print("Total Sales:", total_sales)
print("Average Discount:", avg_discount)
print("Total Products:", total_products)
print("Average Rating:", avg_rating)
```

```python
import plotly.express as px

category_sales = (
    df.groupby('Category')['Sales']
      .sum()
      .reset_index()
      .sort_values('Sales', ascending=False)
)

fig = px.bar(
    category_sales,
    x='Category',
    y='Sales',
    title='Sales by Category',
    color='Sales',
    hover_data={
        'Category': True,
        'Sales': ':,.0f'
    }
)

fig.show()
```

```python
city_sales = (
    df.groupby('City')['Sales']
      .sum()
      .reset_index()
      .sort_values('Sales', ascending=False)
)

fig = px.bar(
    city_sales,
    x='City',
    y='Sales',
    title='Sales by City',
    color='Sales',
    hover_data={
        'City': True,
        'Sales': ':,.0f'
    }
)

fig.show()
```

```python
top_products = (
    df.groupby('ProductName')['Sales']
      .sum()
      .reset_index()
      .sort_values('Sales', ascending=False)
      .head(10)
)

fig = px.bar(
    top_products,
    x='ProductName',
    y='Sales',
    title='Top 10 Products by Sales',
    color='Sales',
    hover_data={
        'ProductName': True,
        'Sales': ':,.0f'
    }
)

fig.show()
```

```python
df['DateAdded'] = pd.to_datetime(df['DateAdded'])

monthly_sales = (
    df.groupby(pd.Grouper(
        key='DateAdded',
        freq='M'
    ))['Sales']
    .sum()
    .reset_index()
)

fig = px.line(
    monthly_sales,
    x='DateAdded',
    y='Sales',
    title='Monthly Sales Trend',
    markers=True,
    hover_data={
        'DateAdded': True,
        'Sales': ':,.0f'
    }
)

fig.show()
```
